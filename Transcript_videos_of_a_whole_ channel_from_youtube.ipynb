```python
output_dir = '/content/gdrive/MyDrive/Kamali_transcriptions'
for video in playlist.videos[:20]:
    try :
      !yt-dlp -x --audio-format mp3 -- {video['link']} # download mp3 from the link
      with open(f"{output_dir}/{video['title']} [{video['id']}].txt", "w") as output_file:
        transcript = pipe(f"{video['title']} [{video['id']}].mp3")['text'] # do the transcription based on the link
        output_file.write(transcript)
    except :
      print('error' ,video['link'])
```
